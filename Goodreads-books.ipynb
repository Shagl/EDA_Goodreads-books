```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
import calendar
```

IMPORT THE DATASET 

```python
books_data= pd.read_csv('books.csv', error_bad_lines=False)
books_data.head(5)
```

DATA EXPLORATION 

```python
books_data.info()
```

```python
books_data.shape
```

MISSING VALUES 

```python
books_data.isnull().sum()
```

DATA PREPROSSING 

```python
books_data['publish_year']= books_data['publication_date'].str.split('/').apply(lambda x:x[2]).astype(int)
books_data['publish_month']= books_data['publication_date'].str.split('/').apply(lambda x:x[0]).astype(int).apply(lambda x: calendar.month_abbr[x])
books_data.head()
```

```python
books_data['average_rating']= round(books_data['average_rating'],1)
books_data.head(5)
```

```python
books_data.rename(columns={'  num_pages': 'num_pages'}, inplace= True)
books_data.drop(columns=["isbn", "isbn13","bookID", "publication_date"],inplace=True)

books_data.head(5)
```
